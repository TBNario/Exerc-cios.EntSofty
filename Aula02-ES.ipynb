# **Parte 1 — Elicitação**

Requisitos Funcionais (RF):

RF01: O sistema deve converter temperaturas de graus Celsius para Fahrenheit.

RF02: O sistema deve converter temperaturas de graus Fahrenheit para Celsius.

RF03: O sistema deve permitir que o usuário insira o valor numérico da temperatura que deseja converter.

Requisitos Não-Funcionais (RNF):

RNF01: O sistema deve ter alta performance, realizando a conversão de forma rápida/imediata.

RNF02: O sistema deve possuir uma usabilidade simples, com uma interface clara para que o usuário escolha a direção da conversão sem confusão.

```python
# ========================================
# CONVERSOR DE TEMPERATURA (DESAFIO BÔNUS)
# Exercício - Engenharia de Software
# ========================================

def celsius_para_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_para_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

def celsius_para_kelvin(celsius):
    return celsius + 273.15

def kelvin_para_celsius(kelvin):
    return kelvin - 273.15

def fahrenheit_para_kelvin(fahrenheit):
    celsius = fahrenheit_para_celsius(fahrenheit)
    return celsius_para_kelvin(celsius)

def kelvin_para_fahrenheit(kelvin):
    celsius = kelvin_para_celsius(kelvin)
    return celsius_para_fahrenheit(celsius)

# ========================================
# PROGRAMA PRINCIPAL
# ========================================

while True: # Loop para várias conversões sem reiniciar
    print("\n" + "="*40)
    print("  CONVERSOR DE TEMPERATURA 🚀")
    print("="*40)
    print("Escolha a conversão:")
    print("1 - Celsius -> Fahrenheit")
    print("2 - Fahrenheit -> Celsius")
    print("3 - Celsius -> Kelvin")
    print("4 - Kelvin -> Celsius")
    print("5 - Fahrenheit -> Kelvin")
    print("6 - Kelvin -> Fahrenheit")
    print("0 - Sair do programa")

    opcao = input("\nOpção: ")

    if opcao == '0':
        print("Saindo do conversor... Valeu, meu nobre!")
        break # Quebra o loop e encerra o programa

    if opcao not in ['1', '2', '3', '4', '5', '6']:
        print("❌ Opção inválida! Tente escolher um número do menu.")
        continue # Volta para o início do loop

    # Validação de entrada: Garante que o usuário digite um número
    while True:
        try:
            valor_str = input("Digite a temperatura: ")
            # Substitui vírgula por ponto caso o usuário digite padrão PT-BR
            valor_str = valor_str.replace(',', '.')
            temp = float(valor_str)
            break # Se a conversão para float der certo, sai do loop de validação
        except ValueError:
            print("❌ Entrada inválida! Por favor, digite apenas números.")

    print("="*40)

    # Realiza a conversão escolhida
    if opcao == '1':
        resultado = celsius_para_fahrenheit(temp)
        print(f"✅ Resultado: {temp:.1f}°C = {resultado:.1f}°F")
    elif opcao == '2':
        resultado = fahrenheit_para_celsius(temp)
        print(f"✅ Resultado: {temp:.1f}°F = {resultado:.1f}°C")
    elif opcao == '3':
        resultado = celsius_para_kelvin(temp)
        print(f"✅ Resultado: {temp:.1f}°C = {resultado:.2f}K")
    elif opcao == '4':
        resultado = kelvin_para_celsius(temp)
        print(f"✅ Resultado: {temp:.2f}K = {resultado:.1f}°C")
    elif opcao == '5':
        resultado = fahrenheit_para_kelvin(temp)
        print(f"✅ Resultado: {temp:.1f}°F = {resultado:.2f}K")
    elif opcao == '6':
        resultado = kelvin_para_fahrenheit(temp)
        print(f"✅ Resultado: {temp:.2f}K = {resultado:.1f}°F")

    print("="*40)
    input("Pressione ENTER para fazer uma nova conversão...")
```
