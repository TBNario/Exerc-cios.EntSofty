# **Parte 1 — Classificação**

Requisitos Funcionais (RFs):

RF01: O sistema deve registrar o nome do exercício.

RF02: O sistema deve validar o peso inserido pelo usuário.

RF03: O sistema deve salvar a quantidade de repetições da série.

Requisitos Não-Funcionais (RNFs):

RNF01: O sistema deve processar o registro em menos de 200ms.

RNF02: O sistema deve proteger os dados dos alunos com criptografia.

RNF03: O sistema deve funcionar com disponibilidade de 99,9% do tempo.

```python
#Parte 2

import time

print("🏋️ GymTrack - Validador de Treino")
print("=" * 40)

# --- DADOS DO TREINO (mude os valores para testar!) ---
exercicio = "Supino Reto"
peso_kg = 80
repeticoes = 10

# ---------------------------------------------------------
# RF01 - O sistema deve validar o nome do exercício
# (não pode ser vazio)
# ---------------------------------------------------------
if exercicio != "":
    print(f"✅ [RF01] Exercício válido: {exercicio}")
else:
    print("❌ [RF01] Erro: O nome do exercício não pode ser vazio.")

# ---------------------------------------------------------
# RF02 - O peso deve estar entre 1 e 300 kg
# ---------------------------------------------------------
if 1 <= peso_kg <= 300:
    print(f"✅ [RF02] Peso válido: {peso_kg}kg")
else:
    print("❌ [RF02] Erro: Peso inválido.")

# ---------------------------------------------------------
# RF03 - As repetições devem estar entre 1 e 50
# ---------------------------------------------------------
if 1 <= repeticoes <= 50:
    print(f"✅ [RF03] Repetições válidas: {repeticoes}")
else:
    print("❌ [RF03] Erro: Repetições inválidas.")

# ---------------------------------------------------------
# RNF01 - O registro deve ocorrer em menos de 200ms
# ---------------------------------------------------------
inicio = time.time()
# Simula o registro no banco de dados
time.sleep(0.05)
print(f"✅ Série registrada: {exercicio} | {peso_kg}kg x {repeticoes} reps")
fim = time.time()

tempo_ms = (fim - inicio) * 1000
if tempo_ms < 200:
    print(f"✅ [RNF01] Tempo de registro: {tempo_ms:.0f}ms ← dentro do limite!")
else:
    print(f"❌ [RNF01] Lento demais: {tempo_ms:.0f}ms ← limite é 200ms")
```

# **REFLEXÃO:**
1. Qual a diferença entre RF e RNF que você percebeu na prática?
R: O RF é a regra de negócio em si (ex: o sistema precisa aceitar um peso entre 1 e 300). Ele define "O QUE" o software faz. Já o RNF define "COMO" o software faz, lidando com a qualidade ou restrição daquela ação (ex: registrar o peso, mas tem que ser rápido, em menos de 200ms).
2. O que aconteceria se esquecêssemos o RNF de performance?
R: O aplicativo funcionaria (salvaria o treino), mas poderia ficar extremamente lento caso o banco de dados demorasse a responder. Isso deixaria os alunos frustrados esperando o loading na tela da academia e eles poderiam parar de usar o app.
