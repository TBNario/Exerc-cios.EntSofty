```python
from dataclasses import dataclass, field
from typing import List
from enum import Enum

#Estrutura Base
class Prioridade(Enum):
    ALTA = "Alta"
    MEDIA = "Média"
    BAIXA = "Baixa"

@dataclass
class RequisitoFuncional:
    id: str
    nome: str
    descricao: str
    prioridade: Prioridade
    ator: str
    pre_condicao: str
    pos_condicao: str

@dataclass
class RequisitoNaoFuncional:
    id: str
    categoria: str
    descricao: str
    criterio_aceitacao: str

#PARTE 2: Função de Validação
def validar_requisito(rf: RequisitoFuncional) -> dict:
    """
    Valida se um requisito funcional segue as boas práticas solicitadas.
    """
    resultados = {
        "Mais de 20 caracteres": len(rf.descricao) > 20,
        "Pré-condição definida": rf.pre_condicao.strip() != "",
        "Critério mensurável (possui números)": any(char.isdigit() for char in rf.descricao)
    }
    return resultados

@dataclass
class SRS:
    projeto: str
    versao: str
    descricao: str
    requisitos_funcionais: List[RequisitoFuncional] = field(default_factory=list)
    requisitos_nao_funcionais: List[RequisitoNaoFuncional] = field(default_factory=list)

    def adicionar_rf(self, req: RequisitoFuncional):
        self.requisitos_funcionais.append(req)
        print(f"✅ RF '{req.id}' adicionado!")

    def adicionar_rnf(self, req: RequisitoNaoFuncional):
        self.requisitos_nao_funcionais.append(req)
        print(f"✅ RNF '{req.id}' adicionado!")

    #PARTE 3: Desafio Extra (Exportar Markdown)
    def exportar_markdown(self):
        md = f"# SRS - {self.projeto}\n\n"
        md += f"**Versão:** {self.versao}\n\n"
        md += f"{self.descricao}\n\n"

        md += "## Requisitos Funcionais\n"
        for rf in self.requisitos_funcionais:
            md += f"### {rf.id} - {rf.nome}\n"
            md += f"- **Descrição:** {rf.descricao}\n"
            md += f"- **Ator:** {rf.ator}\n"
            md += f"- **Pré-condição:** {rf.pre_condicao}\n\n"

        md += "## Requisitos Não-Funcionais\n"
        for rnf in self.requisitos_nao_funcionais:
            md += f"### {rnf.id} ({rnf.categoria})\n"
            md += f"- **Descrição:** {rnf.descricao}\n"
            md += f"- **Critério:** {rnf.criterio_aceitacao}\n\n"

        return md


# PARTE 1: Construir o SRS do "FIAP Marketplace"


fiap_srs = SRS(
    projeto="FIAP Marketplace",
    versao="1.0",
    descricao="Plataforma interna para alunos venderem produtos artesanais entre si."
)

#Adicionando 3 Requisitos Funcionais
fiap_srs.adicionar_rf(RequisitoFuncional(
    id="RF-001",
    nome="Cadastro de Produto",
    descricao="O sistema deve permitir que o aluno cadastre produtos com pelo menos 1 foto e preço.",
    prioridade=Prioridade.ALTA,
    ator="Aluno Vendedor",
    pre_condicao="Usuário logado no sistema",
    pos_condicao="Produto disponível para venda"
))

fiap_srs.adicionar_rf(RequisitoFuncional(
    id="RF-002",
    nome="Busca por Categoria",
    descricao="O usuário deve conseguir filtrar produtos por 5 categorias diferentes.",
    prioridade=Prioridade.MEDIA,
    ator="Aluno Comprador",
    pre_condicao="Existirem produtos cadastrados",
    pos_condicao="Lista de produtos filtrada exibida"
))

fiap_srs.adicionar_rf(RequisitoFuncional(
    id="RF-003",
    nome="Avaliação do Vendedor",
    descricao="O comprador deve poder dar uma nota de 1 a 5 para o vendedor após a compra.",
    prioridade=Prioridade.BAIXA,
    ator="Aluno Comprador",
    pre_condicao="Compra finalizada com sucesso",
    pos_condicao="Média do vendedor atualizada"
))

#Adicionando 2 Requisitos Não-Funcionais
fiap_srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-001",
    categoria="Disponibilidade",
    descricao="O marketplace deve estar online 99.9% do tempo acadêmico.",
    criterio_aceitacao="Monitoramento via UptimeRobot com alerta de queda"
))

fiap_srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-002",
    categoria="Segurança",
    descricao="O sistema deve seguir a LGPD para proteger dados dos alunos.",
    criterio_aceitacao="Criptografia de dados sensíveis e termo de consentimento assinado"
))

#Executando a Validação (Parte 2)
print("\n- Análise Crítica dos Requisitos -")
for rf in fiap_srs.requisitos_funcionais:
    validacao = validar_requisito(rf)
    print(f"Resultado para {rf.id}: {validacao}")

#Gerando Markdown (Parte 3)
print("\n- Gerando Markdown para Notion/Confluence -")
print(fiap_srs.exportar_markdown())
```
